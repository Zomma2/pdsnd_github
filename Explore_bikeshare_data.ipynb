### Question 3

**Most Common User Type**

```r
# Get User Types Value count
table(ny$User.Type)
```

```r
# Get User Types Value count %
table(ny$User.Type)/length(ny$User.Type)*100
```

```r
# Create dataframe for pie chart 
NY_Pie <- data.frame((table(ny$User.Type)/length(ny$User.Type))*100 ,class = c("Unkonwn", "Customer","Subscriber"))
```

```r
# Create only 2 decimal numbers for labeling the pie chart 
count = 0 
Percentage = list()
for (i in table(ny$User.Type)/length(ny$User.Type)*100 ) {
    count = count + 1  
Percentage[count] <- (format(round(i, 2), nsmall = 2)) 
}
```

```r
# Create piechart
ggplot(NY_Pie, aes(x="", y=Freq, fill=class)) +
  geom_bar(stat="identity", width=1, color="white") +
  coord_polar("y", start=0) +
  geom_text(aes(label = Percentage),
            position = position_stack(vjust = 0.5)) + 
ylab('User Type Distribution NY') + 
ggtitle('User Type Pie Chart NY') 
```
